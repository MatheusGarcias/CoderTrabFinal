```python
url ="https://brasilapi.com.br/api/banks/v1"
link = requests.get(url)
link.status_code
```

```python
link.headers
```

```python
link.text
```

```python

from plyer import notification

if link.status_code == 200:
    data_json = link.json()
    notification.notify(
        title='API Request',
        message='A solicitação para a API foi bem-sucedida.',
        app_icon=None,  
        timeout=10  
    )
else:
    print('ERRO AO ACESSAR API')
```

```python
import pandas as pd
import sqlite3
```

```python
db_bancos = 'bancos.db'
df = pd.DataFrame(bancos)
conn = sqlite3.connect(db_bancos)
df.to_sql('bancos', conn, if_exists='replace', index=False)

```

```python
db_cnpj_novo = 'cnpj.db'
df_cnpj = pd.DataFrame(cnpj_novo)
conn = sqlite3.connect(db_bancos)
df_cnpj.to_sql('cnpj', conn, if_exists='replace', index= False)
```
